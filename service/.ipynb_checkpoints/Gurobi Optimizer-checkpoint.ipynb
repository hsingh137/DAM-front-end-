```python
#Insert your code here:
m2 = gp.Model('diet_price')
```

```python
#decision variables
#recipe binary
recipes_var_bin = m2.addVars(price_ingred_df['recipe'].unique().tolist(), vtype = GRB.BINARY)
#ingredients needed 
ingreds = m2.addVars(price_ingred_df['ingreds'].unique().tolist() ,lb = 0)
final_ingreds = m2.addVars(price_ingred_df['ingreds'].unique().tolist() ,lb = 0)
ingreds1 = m2.addVars(price_ingred_df['ingreds'].unique().tolist(),lb = 0)
New_rec_bin=recipes_var_bin = m2.addVars(price_ingred_df['recipe'].unique().tolist(), vtype = GRB.BINARY)
```

```python
price_ingred_df=price_ingred_df.drop_duplicates(['ingred','recipe'])
```

```python
index_tuples = [(name, other) for name in price_ingred_df['recipe'].unique()
                for other in price_ingred_df.loc[price_ingred_df['recipe']== name , 'ingreds'].unique()  ]

multi_ix = pd.MultiIndex.from_tuples(index_tuples)
df_temp = pd.DataFrame(price_ingred_df['stand_quant'].values, index = multi_ix , columns = ['stand_quant'])
recipe_ingredamount = df_temp['stand_quant'].to_dict()
```

```python
ingpred_price_temp = price_ingred_df[['ingreds' , 'Price']].drop_duplicates(keep = 'first')
ingred_price= {
}
for i,row in ingpred_price_temp.iterrows():
    ingred_price[row['ingreds']] = row['Price']
```

```python
#adding recipe nutrionvalues 
# Recommended daily values for different categories of nutrients (ignore the unit)
categories = ['calories', 'Protein', 'fat' , 'Carbohydrate']
minNutrition = {'calories': 9240, #cal
                'Protein': 132, #grams 
                'fat': 30 , #check latererr
                'Carbohydrate': 132 # grams
               }

recipe_nutritionValues = {
}
for i,row in recipes.iterrows():
    if row['name'] in price_ingred_df['recipe'].unique():
        recipe_nutritionValues[(row['name'] , 'calories')] = float(row['calories'])
        recipe_nutritionValues[(row['name'] , 'fat')] = float(row['fat'])
        recipe_nutritionValues[(row['name'] , 'Protein')] = float(row['Protein'])
        recipe_nutritionValues[(row['name'] , 'Carbohydrate')] = float(row['Carbohydrate'])
```

```python
#constraints 

m2.addConstrs(quicksum(recipes_var_bin[r]*recipe_allergenceValue[r,i] for i in allergens_input )==0 for r in price_ingred_df['recipe'].unique())
m2.addConstrs(quicksum(recipes_var_bin[r]*recipe_tagsValue[r,i] for i in tags_input )==0 for r in price_ingred_df['recipe'].unique())
#sum of the recipes needs to be less greater than min nutrition 
recipe_nutrition_const = m2.addConstrs((quicksum(2 * recipe_nutritionValues[f,c] * recipes_var_bin[f] for f in price_ingred_df['recipe'].unique())
              >= minNutrition[c] for c in categories), "Minimum_requirements")

#only 14 recipes
num_of_rec = m2.addConstr(quicksum(recipes_var_bin[f] for f in price_ingred_df['recipe'].unique()) == 7 , 'number_of_recipes')

import math
#ingredient constraint : the totla number of ingredients used across reciepes is que
for i in price_ingred_df['ingred'].unique():
        m2.addConstr(ingreds[i]+fridge[i]  >= (quicksum(recipe_ingredamount[r,i]*recipes_var_bin[r] for r in price_ingred_df['recipe'][price_ingred_df['ingred'] == i].unique())))
```

```python
#objective function 
m2.setObjective(quicksum(ingreds[i]*ingred_price[i] for i in price_ingred_df['ingred'].unique()) , GRB.MINIMIZE)
```
